```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl 

# 2016.01.01 ~ 2021.12.31 추출한 것
url = './open_data_01_2022-05-09.csv'
df = pd.read_csv(url, encoding='cp949')
df.head()
```

```python
# 날짜 잘라주기
for i in range(len(df)):
    df.loc[:]['기준일'][i] = df.loc[:]['기준일'][i][:4]
    
df
```

```python
df_year_drive = df.groupby('기준일')['차량운행'].sum()

df_year_drive
```

```python
df.groupby('기준일')['접수건'].sum()
```

```python
df.groupby('기준일')['접수건'].describe()
```

```python
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
```

```python
ax = df_year_drive.plot(kind='bar', title='연도별 차량운행', figsize=(12,4), legend=True, fontsize=10)
ax.set_xlabel('년도', fontsize=10)
ax.set_ylabel('차량운행', fontsize=10)
ax.legend(['년도','차량운행'], fontsize=10)
```
